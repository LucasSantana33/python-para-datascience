# .loc

- Pegar 1 linha
- Pegar linhas de acordo com alguma condição
- Pegar linhas e colunas específicas
- Pegar 1 valor específico


```python
# pegar uma linha
    display(teste.loc[1:5])
# pegar linhas que correspondem a uma condição
    testes_sublivros_df = teste.loc[teste['Cod.Sub_Livro'] == 283]
    display (testes_sublivros_df)
# pegar 1 valor específico
    print(teste.loc[2, 'Nr.Lote Ativo'])
```
